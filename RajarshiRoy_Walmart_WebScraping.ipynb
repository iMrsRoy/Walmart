```python
#import necessary libraries
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import time
```

```python
#open a browser that we will control
#some required settings to work in Colab
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-div-shm-usage')
driver = webdriver.Chrome('chromedriver', options=options)
```

### Use Selenium to scroll down the page

```python
def makeItScrollable(driver):
    driver.execute_script("return document.getElementsByTagName('html')[0]['style'] = 'overflow: visible'")

def getRidOfOverlay(driver):
    overlay = driver.execute_script("return document.getElementById('YDC-Overlay')")
    if overlay is not None:
        driver.execute_script("return document.getElementById('YDC-Overlay').remove()")

def scrollAllTheWayDown(driver):
    webdriver.ActionChains(driver).send_keys(Keys.END).perform()

def bypassCookiesMessage(driver):
    try:
        acceptAllCookiesButton = driver.find_element(by=By.NAME, value="agree")
    except NoSuchElementException:
        return
    if acceptAllCookiesButton is not None:
        acceptAllCookiesButton.click()
```

```python
#navigate to website with stock data
url = "https://finance.yahoo.com/quote/WMT/history?period1=1577836800&period2=1609372800&interval=1d&filter=history&frequency=1d&includeAdjustedClose=true"
driver.get(url)

#get height of the page
documentHeight = driver.execute_script("return document.documentElement.scrollHeight")

while True:
    print(documentHeight)

    bypassCookiesMessage(driver)
    makeItScrollable(driver)
    getRidOfOverlay(driver)
    scrollAllTheWayDown(driver)

    #wait for data to load
    time.sleep(2)

    #if the page has not gotten taller, stop
    newHeight = driver.execute_script("return document.documentElement.scrollHeight")
    if documentHeight == newHeight:
        break

    #if not, update height
    documentHeight = newHeight
```

## Inspecting Network

```python
import requests
import json
from datetime import date
```

```python
startTime = 1577854800 #Jan 1 2020 00:00:00
endTime = 1609477199 #Dec 31 2020 23:59:59
url = f"https://query1.finance.yahoo.com/v8/finance/chart/WMT?symbol=WMT&period1={startTime}&period2={endTime}&interval=1d"
response = requests.get(url, headers={"user-agent": ""})
jsonStockData = json.loads(response.text)
```

```python
timestamps = jsonStockData["chart"]["result"][0]["timestamp"]
open = jsonStockData["chart"]["result"][0]["indicators"]["quote"][0]["open"]
high = jsonStockData["chart"]["result"][0]["indicators"]["quote"][0]["high"]
low = jsonStockData["chart"]["result"][0]["indicators"]["quote"][0]["low"]
close = jsonStockData["chart"]["result"][0]["indicators"]["quote"][0]["close"]
adjustedClose = jsonStockData["chart"]["result"][0]["indicators"]["adjclose"][0]["adjclose"]
volume = jsonStockData["chart"]["result"][0]["indicators"]["quote"][0]["volume"]
```

```python
print("Date", "Open", "High", "Low", "Close", "Adjusted Close", "Volume", sep="  ")
for t, o, h, l, c, ac, v in zip(timestamps, open, high, low, close, adjustedClose, volume):
    d = date.fromtimestamp(t).strftime("%b %d, %Y")
    print(d, o, h, l, c, ac, v)
```
